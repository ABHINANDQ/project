### 1) Univariate Analysis

```python
# Perform Univariate Analysis using Density plot
data['Engine_Size(L)'].plot.kde()
plt.xlabel('Engine_Size(L)')
plt.tight_layout(pad=0.5)
```

```python
# Perform Univariate Analysis using Density plot
plt.subplot(141)
data['Fuel_Consumption_City(L/100 km)'].plot.kde(figsize=(15,6))
plt.xlabel('Fuel Consumption City (L/100 km)')
plt.tight_layout(pad=0.5)

plt.subplot(142)
data['Fuel_Consumption_Hwy(L/100 km)'].plot.kde()
plt.xlabel('Fuel Consumption Hwy (L/100 km)')
plt.tight_layout(pad=0.5)

plt.subplot(143)
data['Fuel_Consumption_Comb(L/100 km)'].plot.kde()
plt.xlabel('Fuel Consumption Comb (L/100 km)')
plt.tight_layout(pad=0.5)
```

```python
# Perform Univariate Analysis using Density plot
plt.subplot(141)
data['Fuel_Consumption_Comb(mpg)'].plot.kde()
plt.xlabel('Fuel Consumption Comb (mpg)')
plt.tight_layout(pad=0.5)

plt.subplot(142)
data['CO2_Emissions(g/km)'].plot.kde(figsize=(15,6))
plt.xlabel('CO2 Emissions(g/km)')
plt.tight_layout(pad=0.5)

plt.subplot(143)
data['Cylinders'].plot.kde(figsize=(15,6))
plt.xlabel('Cylinders')
plt.tight_layout(pad=0.5)
```

### 2) Bivariate Analysis

```python
data.columns
```

```python
plt.scatter(data["Engine_Size(L)"], data["CO2_Emissions(g/km)"],c="brown")
plt.title('EMISSION')
plt.xlabel('Engine_Size')
plt.ylabel('CO2_Emissions(g/km)')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4,figsize=(16,8)) # create 2x2 array of subplots
data.boxplot(column='Fuel_Consumption_City(L/100 km)',by='Cylinders', ax=axes[0]) # add boxplot to 1st subplot
data.boxplot(column='Fuel_Consumption_Hwy(L/100 km)', by='Cylinders', ax=axes[1]) # add boxplot to 2nd subplot
data.boxplot(column='Fuel_Consumption_Comb(L/100 km)', by='Cylinders', ax=axes[2])
data.boxplot(column='Fuel_Consumption_Comb(mpg)', by='Cylinders', ax=axes[3])
plt.show()
```

```python
# Perform Bivariate Analysis using pairplot
sns.pairplot(
    data,
    x_vars=["Fuel_Consumption_City(L/100 km)","Fuel_Consumption_Hwy(L/100 km)","Fuel_Consumption_Comb(L/100 km)","Fuel_Consumption_Comb(mpg)"],
    y_vars=['CO2_Emissions(g/km)'],
)
```

```python
# Perform Bivariate Analysis using pairplot
sns.pairplot(
    data,
    x_vars=['CO2_Emissions(g/km)'],
    y_vars=['Engine_Size(L)',"Cylinders"],
    
)
```

```python
# Find the correlation and display the heatmap to show the correlation visually
corr_matrix=data.corr()
plt.subplots(figsize=(10,8))
sns.heatmap(corr_matrix,vmin=-1,vmax=1,annot=True)
```
